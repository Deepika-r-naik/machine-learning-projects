My First ML Project

Load data

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

First column is the index number, followed by the other columns; we gonna use the four variables to make a prediction on the LogS variable;

Split the data into X and Y variables

#Data Separation as x and y

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

### Data Splitting - using scikit

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

## Model Building

###Linear Regression


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

**Applying the model to make a prediction**

```python
y_lr_train_prod = lr.predict(x_train)
y_lr_test_prod = lr.predict(x_test)
```

```python
y_lr_test_prod
```

```python
y_lr_test_prod
```

###Evaluate Model Performance

```python
y_train
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_prod)
lr_train_r2 = r2_score(y_train, y_lr_train_prod)

lr_test_mse = mean_squared_error(y_test, y_lr_test_prod)
lr_test_r2 = r2_score(y_test, y_lr_test_prod)
```

```python
print('LR MSE (Train):',lr_train_mse)
print('LR R2 (Train):', lr_train_r2)
print('LR MSE (Test):', lr_test_mse)
print('LR R2 (Test):', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
```

```python
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

##**Random Forest

**Training the model - applying the model to make prediction; using Random Forest regressor model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

```python
y_rf_train_prod = rf.predict(x_train)
y_rf_test_prod = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_prod)
rf_train_r2 = r2_score(y_train, y_rf_train_prod)

rf_test_mse = mean_squared_error(y_test, y_rf_test_prod)
rf_test_r2 = r2_score(y_test, y_rf_test_prod)
```

```python
rf_results = pd.DataFrame(['Linear regression', rf_train_mse, rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## Model Comparision

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```
